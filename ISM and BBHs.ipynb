```python
# Define the constants
G = 6.67430e-11  # Gravitational constant in m^3 kg^(-1) s^(-2)
msun = 1.99*1e30
m = 10*msun  # Mass constant
m_H = 1.67*1e-27
rho = 1*m_H / (1e-2)**3  # Density constant
delta_t = 1e9
M = 2*m
r_s = 696340*1e3 #meters
c = 3e8

# Define the function for dr/dt in terms of r and constants
def dr_dt_DF(r):
    r = r*r_s # converting to meters
    v = np.sqrt(2*G*M/r)
    I = (v**3 * delta_t /(2*G*m))  
    return 2 * 8 * np.pi * (G ** 0.5) * m * rho * I * (r ** (5 / 2)) / (M ** (3 / 2))  # the intial 2 is to 
                                                                                #accomodate for both black holes

def dr_dt_AMC(r):
    r = r*r_s # converting to meters
    c_s = sqrt(1 * (1.38e-23) * (100) / m_H)
    yr = 365.25 * 24 * 3600
    alpha = 4 * np.pi * 1.12 * G ** 2 * rho / c_s ** 3
    alpha_m = alpha * m
    t = 1e0 * yr
    
    deta_1_dt = alpha_m/(alpha_m*t - 1)**2
    eta_1 = -1/(alpha_m*t-1)
    return 3 * deta_1_dt / eta_1 * r # assuming that both black holes are equal mass

def dr_dt_GW(r):
    r = r*r_s # converting to meters
    mu = m**2/ (2*m)
    return 64 * G**3 * mu * M**2 / (5 * c**5 * r**3)

# Generate values for r
r_values = np.linspace(1, 300, 100)  # You can adjust the range and number of points


# ----------------   Calculate corresponding values for dr/dt   ----------------
dr_dt_values = dr_dt_DF(r_values)
# Plot the curve for dynamical friction
plt.plot(r_values, dr_dt_values, label= 'Dynamical friction')


#  ----------------  Calculate corresponding values for dr/dt   ----------------
dr_dt_values = dr_dt_AMC(r_values)
# Plot the curve for decay due to angular momentum conservation
plt.plot(r_values, dr_dt_values, label= 'Isotropic accretion')


#  ----------------  Calculate corresponding values for dr/dt   ----------------
dr_dt_values = dr_dt_GW(r_values)
# Plot the curve for decay due to angular momentum conservation
plt.plot(r_values, dr_dt_values, label= 'GW emission')


# -------------------------------------
# now for a different mass 
# -------------------------------------

m = 30*msun
M = 2*m

# ----------------   Calculate corresponding values for dr/dt   ----------------
dr_dt_values = dr_dt_DF(r_values)
# Plot the curve for dynamical friction
plt.plot(r_values, dr_dt_values, linestyle = '--', color = colors[0])


#  ----------------  Calculate corresponding values for dr/dt   ----------------
dr_dt_values = dr_dt_AMC(r_values)
# Plot the curve for decay due to angular momentum conservation
plt.plot(r_values, dr_dt_values, linestyle = '--', color = colors[1])


#  ----------------  Calculate corresponding values for dr/dt   ----------------
dr_dt_values = dr_dt_GW(r_values)
# Plot the curve for decay due to angular momentum conservation
plt.plot(r_values, dr_dt_values,  linestyle = '--', color = colors[2])

# Add a legend
legend1 = plt.legend(loc = 'best', fontsize = 10)

# Additional legend
legend_elements = [
    plt.Line2D([0], [0], color='black', linestyle='-', label='$m_{\\rm BH} = 10 \\, M_\\odot$'),
    plt.Line2D([0], [0], color='black', linestyle='--', label='$m_{\\rm BH} = 30 \\, M_\\odot$')
]

# Get the current axes
ax = plt.gca()
# Add the additional legend to the axes
legend2 = ax.legend(handles=legend_elements,  bbox_to_anchor=(1.08, 1.2), ncol = 2)
# Add the existing legend back to the axes
ax.add_artist(legend1)

plt.xlabel('$r \,\, [R_\odot]$')
plt.ylabel(r'$|\dot{r}|$ [ms$^{-1}]$')

plt.yscale('log')
plt.ylim(2e-13,5e-2)
plt.locator_params(axis='y', numticks=6)  # Adjust numticks as needed
plt.locator_params(axis='x', nbins=5)
# plt.savefig('figures/rate_of_orbital_decay')
```
